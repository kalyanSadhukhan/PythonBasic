```python
import pandas as pd

# Load dataset
df = pd.read_csv("digital_marketing_campaign_dataset.csv")

# Check for missing values
print(df.isnull().sum())

# Fill missing numerical values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing categorical values with mode
df.fillna(df.mode().iloc[0], inplace=True)

# Verify data types
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Ad Spend
plt.figure(figsize=(8,5))
sns.histplot(df['AdSpend'], bins=50, kde=True)
plt.title("Distribution of Ad Spend")
plt.xlabel("Ad Spend")
plt.show()

# Boxplot of Conversion Rate by Campaign Type
plt.figure(figsize=(8,5))
sns.boxplot(x=df['CampaignType'], y=df['ConversionRate'])
plt.title("Conversion Rate by Campaign Type")
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.stats import ttest_ind

# Compare conversion rates of Social Media and PPC campaigns
social_media_conversion = df[df['CampaignChannel'] == "Social Media"]['ConversionRate']
ppc_conversion = df[df['CampaignChannel'] == "PPC"]['ConversionRate']

t_stat, p_value = ttest_ind(social_media_conversion, ppc_conversion)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
# Heatmap to show correlation
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd

# Load dataset
df = pd.read_csv("digital_marketing_campaign_dataset.csv")

# Create dashboard app
app = dash.Dash(__name__)

# Ad Spend vs. Conversion Rate Scatterplot
scatter_fig = px.scatter(df, x="AdSpend", y="ConversionRate",
                         color="CampaignChannel",
                         title="Ad Spend vs. Conversion Rate")

# Campaign Type vs. Conversion Rate Boxplot
boxplot_fig = px.box(df, x="CampaignType", y="ConversionRate",
                     title="Conversion Rate by Campaign Type")

# Layout
app.layout = html.Div(children=[
    html.H1("Marketing Analytics Dashboard"),

    dcc.Graph(id="scatter_plot", figure=scatter_fig),

    dcc.Graph(id="boxplot", figure=boxplot_fig)
])

if __name__ == '__main__':
    app.run(debug=True)
```
