```python
import pandas as pd

# Load dataset
df = pd.read_csv("Salary Data.csv")

# Check for missing values
print(df.isnull().sum())

# Fill missing salary values with median
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Fill missing education levels with mode
df['Education Level'] = df['Education Level'].fillna(df['Education Level'].mode()[0])
# Verify data types
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatterplot of Salary vs. Years of Experience
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Years of Experience'], y=df['Salary'])
plt.title("Salary vs. Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

# Boxplot of Salary by Education Level
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Education Level'], y=df['Salary'])
plt.title("Salary by Education Level")
plt.show()
```

```python
from scipy.stats import ttest_ind

# Compare salaries of two education groups
bachelor_salaries = df[df['Education Level'] == "Bachelor's"]['Salary']
masters_salaries = df[df['Education Level'] == "Master's"]['Salary']

t_stat, p_value = ttest_ind(bachelor_salaries, masters_salaries)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Heatmap to show correlation
plt.figure(figsize=(8,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
